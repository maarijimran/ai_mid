```python
# Graph
cave = {
    'Entry': ['TunnelA', 'TunnelB'],
    'TunnelA': ['Entry', 'ChamberX'],
    'TunnelB': ['Entry', 'ChamberY'],
    'ChamberX': ['TunnelA', 'DeepCave'],
    'ChamberY': ['TunnelB'],
    'DeepCave': ['ChamberX']
}

ore_nodes = ['ChamberY', 'DeepCave']
```

```python
def dfsTraversal(node, goal, path = None, visited = None):
    if visited == None:
        visited = set()
    if path == None:
        path = [node]

    if node == goal:
        return path

    visited.add(node)

    for neighbor in cave[node]:
        if neighbor not in visited:
            result = dfsTraversal(neighbor, goal, path + [neighbor], visited)
            if result:
                return result

    return None
```

```python
def calcUtility(order):
    totalUtility = 0
    details = []
    current = 'Entry'
    totalMoves = 0
    oresCollected = 0

    for target in order:
        path = dfsTraversal(current, target)
        moves = len(path) - 1
        
        totalMoves += moves
        oresCollected += 1

        details.append((current, target, moves, path))
        current = target

    totalUtility = (10 * oresCollected) - totalMoves

    return totalUtility, details
```

```python
import itertools

def findOres():
    bestUtility = float('-inf')
    bestPath = None

    for order in itertools.permutations(ore_nodes):
        utility, details = calcUtility(order)

        if utility > bestUtility:
            bestUtility = utility
            bestPath = (order, details)

    print("Best Order:", bestPath[0])
    print()

    for current, target, moves, path in bestPath[1]:
        print(f"Path to {target}: {path}")
        print(f"Moves: {moves}")
        print()

    print("Total Utility:", bestUtility)
```

```python
findOres()
```
