```python
from collections import deque

# -------------------
# Hospital Graph
# -------------------
hospital = {
    "WardA": {"status": "Critical", "adjacent": ["WardB", "WardC"]},
    "WardB": {"status": "Normal", "adjacent": ["WardA", "WardD"]},
    "WardC": {"status": "Critical", "adjacent": ["WardA", "WardD"]},
    "WardD": {"status": "Empty", "adjacent": ["WardB", "WardC"]}
}

# -------------------
# BFS → nearest critical ward
# -------------------
def bfs(start):
    queue = deque([(start, [start])])
    visited = set()

    while queue:
        node, path = queue.popleft()

        # If found a critical ward (excluding current if already treated)
        if hospital[node]["status"] == "Critical":
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in hospital[node]["adjacent"]:
                queue.append((neighbor, path + [neighbor]))

    return None

# -------------------
# Utility-Based Agent
# -------------------
class UtilityBasedAgent:
    def __init__(self, start):
        self.current = start
        self.moves = 0
        self.treated = 0

    def utility(self):
        return (10 * self.treated) - self.moves

    def act(self):
        # If current ward is critical → treat
        if hospital[self.current]["status"] == "Critical":
            print(f"At {self.current}: Treating patient")
            hospital[self.current]["status"] = "Treated"
            self.treated += 1
            print(f"Utility: {self.utility()}\n")
            return

        # Otherwise → find nearest critical ward
        path = bfs(self.current)

        if path is None:
            return False  # no more critical patients

        # Move one step along path
        next_node = path[1]
        print(f"Moving from {self.current} → {next_node}")

        self.current = next_node
        self.moves += 1

        print(f"Utility: {self.utility()}\n")
        return True

# -------------------
# Simulation
# -------------------
def run_simulation():
    agent = UtilityBasedAgent("WardB")  # start anywhere (you can change)

    print("Starting Simulation...\n")

    while True:
        # Check if any critical patients remain
        remaining = any(
            hospital[w]["status"] == "Critical" for w in hospital
        )

        if not remaining:
            break

        cont = agent.act()
        if cont is False:
            break

    print("Final Utility:", agent.utility())

# -------------------
# Run
# -------------------
run_simulation()
```
