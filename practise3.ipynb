```python
import itertools
from collections import deque

# Utility = (15×urgent + 5×normal) - moves.

warehouse = {
'Start':['A','B'],
'A':['Start','C','D'],
'B':['Start','E'],
'C':['A'],
'D':['A'],
'E':['B']
}
packages = {'C':'URGENT','D':'NORMAL','E':'URGENT'}

def bfs(start, goal):
    visited = set()
    queue = deque([(start, [start])])
    visited.add(start)

    while queue:
        node, path = queue.popleft()

        if node == goal:
            return path

        for neighbor in warehouse[node]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))
                visited.add(neighbor)

    return None

def calcUtility(order):
    totalUtility = 0
    urgent = 0
    normal = 0
    totalMoves = 0
    current = 'Start'
    details = []

    for target in order:
        path = bfs(current, target)
        moves = len(path) - 1
        totalMoves += moves
        details.append((target, path, moves))
        
        if packages[target] == "URGENT":
            urgent += 1
        else:
            normal += 1

        current = target

    totalUtility = (15 * urgent + normal * 5) - totalMoves
    return totalUtility, details

def track():
    best = float('-inf')
    bestPath = None

    for order in itertools.permutations(packages):
        utility, path = calcUtility(order)

        if utility > best:
            best = utility
            bestPath = (order, path)

    print("best order : ", bestPath[0])
    print()

    for target, path, moves in bestPath[1]:
        print(f"Path to target : {path}")
        print(f"Moves : {moves}")

    print(f"total Utility : {best}")
        
```

```python
track()
```
