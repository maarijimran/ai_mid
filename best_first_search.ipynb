```python
from queue import PriorityQueue

# Example graph represented as an adjacency list
graph = {
    'A': [('B', 5), ('C', 8)],
    'B': [('D', 10)],
    'C': [('E', 3)],
    'D': [('F', 7)],
    'E': [('F', 2)],
    'F': []
}
```

```python
def bestFitSearch(start, goal):
    visited = set()
    pq = PriorityQueue()
    pq.put((0, start))

    while not pq.empty():
        cost, node = pq.get()

        if node not in visited:
            print(node, end=' ')
            visited.add(node)

            if node == goal:
                return print("Goal Founded")

            for neighbor, weight in graph[node]:
                if neighbor not in visited:
                    pq.put((weight, neighbor))
    
    return print("Not Founded")
```

```python
bestFitSearch('A', 'F')
```
