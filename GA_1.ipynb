```python
import random
import math

# Q7 — GA with PMX
# Solve TSP using:
# - Roulette Selection
# - PMX crossover
# - Swap mutation
# cities = {
# 0:(0,0),1:(3,4),2:(6,1),
# 3:(7,5),4:(2,7),5:(5,3)
# }

# Cities
cities = {
    0:(0,0), 1:(3,4), 2:(6,1),
    3:(7,5), 4:(2,7), 5:(5,3)
}

# ----------------------------
# Distance Functions
# ----------------------------
def distance(a, b):
    x1, y1 = cities[a]
    x2, y2 = cities[b]
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)

def total_distance(route):
    dist = 0
    for i in range(len(route)-1):
        dist += distance(route[i], route[i+1])
    dist += distance(route[-1], route[0])  # return to start
    return dist

# ----------------------------
# Initial Population
# ----------------------------
def init_population(size):
    population = []
    nodes = list(cities.keys())
    for _ in range(size):
        ind = nodes[:]
        random.shuffle(ind)
        population.append(ind)
    return population

# ----------------------------
# Roulette Selection
# ----------------------------
def roulette_selection(pop):
    fitness = [1/total_distance(ind) for ind in pop]
    total_fit = sum(fitness)
    
    probs = [f/total_fit for f in fitness]
    parents = random.choices(pop, weights=probs, k=2)
    
    return parents[0], parents[1]

# ----------------------------
# PMX Crossover
# ----------------------------
def pmx(p1, p2):
    size = len(p1)
    c1 = [None]*size
    c2 = [None]*size

    # crossover points
    a = random.randint(0, size-2)
    b = random.randint(a+1, size-1)

    # copy segment
    c1[a:b+1] = p1[a:b+1]
    c2[a:b+1] = p2[a:b+1]

    def fill(child, parent):
        for i in range(size):
            if child[i] is None:
                gene = parent[i]
                while gene in child:
                    gene = parent[child.index(gene)]
                child[i] = gene
        return child

    c1 = fill(c1, p2)
    c2 = fill(c2, p1)

    return c1, c2

# ----------------------------
# Swap Mutation
# ----------------------------
def swap_mutation(ind, rate=0.1):
    for i in range(len(ind)):
        if random.random() < rate:
            j = random.randint(0, len(ind)-1)
            ind[i], ind[j] = ind[j], ind[i]
    return ind

# ----------------------------
# Genetic Algorithm
# ----------------------------
def genetic_algorithm(pop_size=10, generations=50):
    population = init_population(pop_size)

    best = min(population, key=total_distance)
    best_dist = total_distance(best)

    for gen in range(generations):
        new_pop = []

        while len(new_pop) < pop_size:
            p1, p2 = roulette_selection(population)

            c1, c2 = pmx(p1, p2)

            c1 = swap_mutation(c1)
            c2 = swap_mutation(c2)

            c2 = swap_mutation(c2)
            new_pop.extend([c1, c2])

        population = new_pop[:pop_size]

        # update best
        current_best = min(population, key=total_distance)
        current_dist = total_distance(current_best)

        if current_dist < best_dist:
            best = current_best
            best_dist = current_dist

        print(f"Gen {gen+1}: Best Distance = {best_dist:.2f}")

    return best, best_dist

# ----------------------------
# Run
# ----------------------------
best_route, best_distance = genetic_algorithm()

print("\nBest Route:", best_route)
print("Best Distance:", best_distance)
```
