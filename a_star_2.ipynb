```python
from queue import PriorityQueue
```

```python
grid = [
    ['S', '.', '.', '#', '.'],
    ['.', 'C', '.', '#', '.'],
    ['.', '#', '.', '.', '.'],
    ['.', '.', 'T', '.', '.'],
    ['#', '.', '.', '.', 'G']
]

rows = len(grid)
cols = len(grid[0])

for i in range(rows):
    for j in range(cols):
        if grid[i][j] == 'S':
            start = (i, j)
        if grid[i][j] == 'G':
            goal = (i, j)

print(f"{start}, {goal}")
```

```python
def get_cost(cell):
    if cell == '.':
        return 1
    elif cell == 'C':
        return 4
    elif cell == 'T':
        return 6
    else:
        return float('inf')
```

```python
def heuristic(a, b):
    (x1, y1) = a
    (x2, y2) = b
    return abs(x1 - x2) + abs(y1 - y2)
```

```python
def a_star(start, goal):
    visited = set()
    pq = PriorityQueue()
    pq.put((heuristic(start, goal), start))
    track = {start : None}
    g_goal = {start : 0}

    while not pq.empty():
        f, node = pq.get()

        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = track[node]
            path.reverse()
            return path

        if node in visited:
            continue

        visited.add(node)
        x, y = node

        for dx, dy in [(1,0),(-1,0),(0,1),(0,-1)]:
            nx, ny = dx + x, dy + y
            if 0 <= nx < rows and 0 <= ny < cols:
                if grid[nx][ny] != '#':
                    new_cost = g_goal[node] + get_cost(grid[nx][ny])
                    if (nx,ny) not in g_goal or g_goal[(nx,ny)] > new_cost:
                        f = new_cost + heuristic((nx,ny), goal)
                        g_goal[(nx,ny)] = new_cost
                        pq.put((f, (nx, ny)))
                        track[(nx,ny)] = node

    return None     
```

```python
path = a_star(start, goal)
```
