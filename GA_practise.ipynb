```python
import math
import random

# Cities
cities = {
    0:(0,0), 1:(3,4), 2:(6,1),
    3:(7,5), 4:(2,7), 5:(5,3)
}

def distance(a,b):
    x1, y1 = cities[a]
    x2, y2 = cities[b]
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)

def total_distance(route):
    size = len(route)
    total_distance = 0
    for i in range(size - 1):
        total_distance += distance(route[i], route[i + 1])
    total_distance += distance(route[-1], route[0])
    return total_distance

def init_population(size):
    population = []
    node = list(cities.keys())
    for _ in range(size):
        ind = node[:]
        random.shuffle(ind)
        population.append(ind)
    return population

def roulette_selection(pop):
    fitness = [1/total_distance(ind) for ind in pop]
    total_fitness = sum(fitness)

    probs = [f/total_fitness for f in fitness]
    parents = random.choices(pop, weights=probs, k=2)

    return parents[0], parents[1]

def pmx(p1, p2):
    size = len(p1)
    c1 = [None] * size
    c2 = [None] * size

    a = random.randint(0, size - 2)
    b = random.randint(a + 1, size - 1)

    c1[a:b+1] = p1[a:b+1]
    c2[a:b+1] = p2[a:b+1]

    def fill(c, p):
        for i in range(len(c)):
            if c[i] == None:
                gen = p[i]
                while gen in c:
                    gen = p[c.index(gen)]
                c[i] = gen
        return c
    
    c1 = fill(c1, p2)
    c2 = fill(c2, p1)

    return c1, c2

def swap_mutation(ind, rate=0.1):
    for i in range(len(ind)):
        if random.random() < rate:
            j = random.randint(0, len(ind) - 1)
            ind[i], ind[j] = ind[j], ind[i]
    return ind

def genetic_algo(pop_size=15, generation=100):
    population = init_population(pop_size)
    best = min(population, key=total_distance)
    best_dist = total_distance(best)

    for gen in range(generation):
        new_pop = []

        while len(new_pop) < len(population):
            p1, p2 = roulette_selection(population)
            c1, c2 = pmx(p1, p2)

            c1 = swap_mutation(c1, 1000000000)
            c2 = swap_mutation(c2, 1000000000)
            new_pop.extend([c1, c2])

        population = new_pop[:pop_size]

        current_best = min(population, key=total_distance)
        curr_distance = total_distance(current_best)

        if curr_distance < best_dist:
            best_dist = curr_distance
            best = current_best

        print(f"Gen {gen+1}: Best Distance = {best_dist:.2f}")

    return best, best_dist

best, best_dist = genetic_algo(80, 100)
print("\nBest Route:", best)
print("Best Distance:", best_dist)
```
