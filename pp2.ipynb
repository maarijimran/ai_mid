```python
from queue import PriorityQueue

# -------------------
# Correct Grid (From Question)
# -------------------
grid = [
    ['S', '.', '.', 'W', '.'],
    ['.', '#', '.', '#', '.'],
    ['.', '#', '.', '.', '.'],
    ['.', '.', 'F', '#', '.'],
    ['.', '.', '.', '.', 'G']
]

rows, cols = len(grid), len(grid[0])

# -------------------
# Find Start & Goal
# -------------------
for i in range(rows):
    for j in range(cols):
        if grid[i][j] == 'S':
            start = (i, j)
        if grid[i][j] == 'G':
            goal = (i, j)

# -------------------
# Cost Function
# -------------------
def get_cost(cell):
    if cell == '.' or cell == 'S' or cell == 'G':
        return 1
    elif cell == 'W':  # Water
        return 3
    elif cell == 'F':  # Fire
        return 5
    else:
        return float('inf')  # blocked

# -------------------
# Manhattan Heuristic
# -------------------
def heuristic(pos):
    return abs(pos[0] - goal[0]) + abs(pos[1] - goal[1])

# -------------------
# A* Algorithm
# -------------------
def a_star(start, goal):
    visited = set()
    pq = PriorityQueue()
    
    pq.put((heuristic(start), start))  # f(n), node
    track = {start: None}
    g_cost = {start: 0}

    expanded_nodes = []  # for step tracking

    while not pq.empty():
        f, node = pq.get()

        if node in visited:
            continue

        visited.add(node)
        expanded_nodes.append(node)

        # Goal check
        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = track[node]
            return path[::-1], expanded_nodes, g_cost[goal]
        
        x, y = node

        # 4-direction movement
        for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:
            nx, ny = x + dx, y + dy

            if 0 <= nx < rows and 0 <= ny < cols:
                if grid[nx][ny] == '#':
                    continue  # blocked

                new_g = g_cost[node] + get_cost(grid[nx][ny])

                if (nx, ny) not in g_cost or new_g < g_cost[(nx, ny)]:
                    g_cost[(nx, ny)] = new_g
                    f_value = new_g + heuristic((nx, ny))

                    pq.put((f_value, (nx, ny)))
                    track[(nx, ny)] = node

    return None, expanded_nodes, None

# -------------------
# Run
# -------------------
path, expanded, cost = a_star(start, goal)

print("Expanded Nodes:", expanded)
print("Optimal Path:", path)
print("Total Cost:", cost)
```
