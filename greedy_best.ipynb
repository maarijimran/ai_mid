```python
# Graph with different edge costs
graph = {
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {},
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {}
}

# Heuristic function (estimated cost to reach goal 'I')
heuristic = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 4,
    'E': 7,
    'F': 3,
    'G': 6,
    'H': 2,
    'I': 0  # Goal node
}

from queue import PriorityQueue

def greedy_best(start, goal):
    visited = set()
    pq = PriorityQueue()
    pq.put((heuristic[start], start))
    track = {start : None}

    while not pq.empty():
        cost, node = pq.get()

        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = track[node]
            path.reverse()
            return path
        
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                pq.put((heuristic[neighbor], neighbor))
                track[neighbor] = node
    
    return None

path = greedy_best('A', 'I')
print(path)
```

```python
from queue import PriorityQueue

grid = [
    ['S', '.', 'D', '#', '.'],
    ['.', '#', '.', 'R', '.'],
    ['.', 'D', '.', '#', '.'],
    ['R', '.', '.', '.', '.'],
    ['#', '.', 'R', '.', 'G']
]

rows, cols = len(grid), len(grid[0])

# Find start and goal
for i in range(rows):
    for j in range(cols):
        if grid[i][j] == 'S':
            start = (i, j)
        if grid[i][j] == 'G':
            goal = (i, j)

# Manhattan heuristic
def heuristic(pos):
    return abs(pos[0] - goal[0]) + abs(pos[1] - goal[1])

def greedy_bfs(start, goal):
    visited = set()
    pq = PriorityQueue()
    
    pq.put((heuristic(start), start))
    track = {start: None}

    while not pq.empty():
        h, node = pq.get()

        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = track[node]
            return path[::-1]
        
        if node in visited:
            continue
        
        visited.add(node)

        x, y = node

        # Same neighbor expansion
        for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:
            nx, ny = x + dx, y + dy

            if (0 <= nx < rows and 
                0 <= ny < cols and 
                grid[nx][ny] != '#'):

                if (nx, ny) not in visited:
                    pq.put((heuristic((nx, ny)), (nx, ny)))
                    track[(nx, ny)] = node

    return None

# Run
path = greedy_bfs(start, goal)
print("Path:", path)
```
