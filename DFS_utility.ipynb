```python
# Graph
cave = {
    'Entry': ['TunnelA', 'TunnelB'],
    'TunnelA': ['Entry', 'ChamberX'],
    'TunnelB': ['Entry', 'ChamberY'],
    'ChamberX': ['TunnelA', 'DeepCave'],
    'ChamberY': ['TunnelB'],
    'DeepCave': ['ChamberX']
}

ore_nodes = ['ChamberY', 'DeepCave']


# DFS to find a path
def dfs(start, goal, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = [start]

    # Goal reached
    if start == goal:
        return path

    visited.add(start)

    for neighbor in cave[start]:
        if neighbor not in visited:
            result = dfs(neighbor, goal, visited, path + [neighbor])
            if result:
                return result

    return None


# Compute utility for a given order
def compute_utility(order):
    current = 'Entry'
    total_moves = 0
    ores_collected = 0
    details = []

    for target in order:
        path = dfs(current, target)
        moves = len(path) - 1

        total_moves += moves
        ores_collected += 1

        details.append((target, path, moves))
        current = target

    utility = (10 * ores_collected) - total_moves
    return utility, details


# Main agent (tries all orders)
import itertools

def ore_agent():
    best_utility = float('-inf')
    best_plan = None

    for order in itertools.permutations(ore_nodes):
        utility, details = compute_utility(order)

        if utility > best_utility:
            best_utility = utility
            best_plan = (order, details)

    # Print result
    print("Best Order:", best_plan[0])
    print()

    for target, path, moves in best_plan[1]:
        print(f"Path to {target}: {path}")
        print(f"Moves: {moves}")
        print()

    print("Total Utility:", best_utility)


# Run
ore_agent()
```
