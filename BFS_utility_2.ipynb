```python
from collections import deque
import itertools

# Graph
warehouse = {
    'Start': ['A', 'B'],
    'A': ['Start', 'C', 'D'],
    'B': ['Start', 'E'],
    'C': ['A'],
    'D': ['A'],
    'E': ['B']
}

# Package priorities
packages = {'C': 'URGENT', 'D': 'NORMAL', 'E': 'URGENT'}

# BFS for shortest path
def bfs(start, goal):
    queue = deque([(start, [start])])
    visited = set()

    while queue:
        node, path = queue.popleft()

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in warehouse[node]:
                queue.append((neighbor, path + [neighbor]))

    return None


# Compute utility for a given order
def compute_utility(order):
    current = 'Start'
    total_moves = 0
    urgent = 0
    normal = 0
    details = []

    for target in order:
        path = bfs(current, target)
        moves = len(path) - 1

        total_moves += moves

        if packages[target] == 'URGENT':
            urgent += 1
        else:
            normal += 1

        details.append((target, path, moves))
        current = target

    utility = (15 * urgent + 5 * normal) - total_moves
    return utility, details


# Main agent
def warehouse_agent():
    best_utility = float('-inf')
    best_plan = None

    nodes = list(packages.keys())

    for order in itertools.permutations(nodes):
        utility, details = compute_utility(order)

        if utility > best_utility:
            best_utility = utility
            best_plan = (order, details)

    # Print results
    print("Best Order:", best_plan[0])
    print()

    for target, path, moves in best_plan[1]:
        print(f"Path to {target}: {path}")
        print(f"Moves: {moves}")
        print()

    print("Total Utility:", best_utility)


# Run
warehouse_agent()
```
