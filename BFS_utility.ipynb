```python
from collections import deque
import itertools

# Graph
city = {
    'HQ': ['Plaza', 'Market'],
    'Plaza': ['HQ', 'School'],
    'Market': ['HQ', 'Hospital'],
    'School': ['Plaza', 'Park'],
    'Hospital': ['Market'],
    'Park': ['School']
}

injured = ['School', 'Hospital']

# BFS function to find shortest path
def bfs(start, goal):
    queue = deque([(start, [start])])
    visited = set()

    while queue:
        node, path = queue.popleft()

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in city[node]:
                queue.append((neighbor, path + [neighbor]))

    return None


# Function to compute utility for a given order
def compute_utility(order):
    current = 'HQ'
    total_utility = 0
    full_path = []

    for target in order:
        path = bfs(current, target)
        moves = len(path) - 1
        utility = 10 - moves

        total_utility += utility
        full_path.append((target, path, moves, utility))

        current = target

    return total_utility, full_path


# Utility-based agent (tries all orders)
def rescue_agent():
    best_utility = float('-inf')
    best_plan = None

    # Try all possible rescue orders
    for order in itertools.permutations(injured):
        utility, path_details = compute_utility(order)

        if utility > best_utility:
            best_utility = utility
            best_plan = (order, path_details)

    # Print best result
    print("Best Rescue Order:", best_plan[0])
    print()

    for target, path, moves, utility in best_plan[1]:
        print(f"Path to {target}: {path}")
        print(f"Moves: {moves}, Utility: {utility}")
        print()

    print("Total Utility:", best_utility)


# Run agent
rescue_agent()
```
