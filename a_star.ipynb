```python
from queue import PriorityQueue

grid = [
    ['S', '.', '.', '#', '.'],
    ['.', 'C', '.', '#', '.'],
    ['.', '#', '.', '.', '.'],
    ['.', '.', 'T', '.', '.'],
    ['#', '.', '.', '.', 'G']
]

rows, cols = len(grid), len(grid[0])

# Find start and goal
for i in range(rows):
    for j in range(cols):
        if grid[i][j] == 'S':
            start = (i, j)
        if grid[i][j] == 'G':
            goal = (i, j)

# Cost function
def get_cost(cell):
    if cell == '.':
        return 1
    elif cell == 'C':
        return 4
    elif cell == 'T':
        return 6
    else:
        return float('inf')  # blocked

# Manhattan heuristic
def heuristic(pos):
    return abs(pos[0] - goal[0]) + abs(pos[1] - goal[1])

def a_star(start, goal):
    visited = set()
    pq = PriorityQueue()
    
    pq.put((heuristic(start), start))  # f(n), node
    track = {start: None}
    g_cost = {start: 0}

    while not pq.empty():
        f, node = pq.get()

        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = track[node]
            return path[::-1]
        
        if node in visited:
            continue
        
        visited.add(node)

        x, y = node

        # Same structure as your graph loop
        for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:
            nx, ny = x + dx, y + dy

            if 0 <= nx < rows and 0 <= ny < cols:
                if grid[nx][ny] == '#':
                    continue  # blocked

                new_g = g_cost[node] + get_cost(grid[nx][ny])

                if (nx, ny) not in g_cost or new_g < g_cost[(nx, ny)]:
                    g_cost[(nx, ny)] = new_g
                    f_value = new_g + heuristic((nx, ny))

                    pq.put((f_value, (nx, ny)))
                    track[(nx, ny)] = node

    return None

# Run
path = a_star(start, goal)
print("Path:", path)
```

```python
from queue import PriorityQueue
import math

# Q6 — A* with Euclidean Heuristic
# Use Euclidean Distance heuristic.
# Grid:
# S . . # .
# . G_ . # .
# . # . S_ .
# . . . # .
# # . G_ . G

rows, cols = len(grid), len(grid[0])

# Find start and goal
for i in range(rows):
    for j in range(cols):
        if grid[i][j] == 'S':
            start = (i, j)
        if grid[i][j] == 'G':
            goal = (i, j)

# Cost function: all cells cost 1 (except blocked)
def get_cost(cell):
    if cell == '#':
        return float('inf')
    else:
        return 1

# Euclidean heuristic
def heuristic(pos):
    return math.sqrt((pos[0] - goal[0])**2 + (pos[1] - goal[1])**2)

def a_star(start, goal):
    visited = set()
    pq = PriorityQueue()
    pq.put((heuristic(start), start))
    
    track = {start: None}
    g_cost = {start: 0}

    while not pq.empty():
        f, node = pq.get()

        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = track[node]
            return path[::-1]
        
        if node in visited:
            continue
        
        visited.add(node)
        x, y = node

        # Expand neighbors (4 directions)
        for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:
            nx, ny = x + dx, y + dy

            if 0 <= nx < rows and 0 <= ny < cols and grid[nx][ny] != '#':
                new_g = g_cost[node] + get_cost(grid[nx][ny])

                if (nx, ny) not in g_cost or new_g < g_cost[(nx, ny)]:
                    g_cost[(nx, ny)] = new_g
                    f_value = new_g + heuristic((nx, ny))
                    pq.put((f_value, (nx, ny)))
                    track[(nx, ny)] = node

    return None

# Run
path = a_star(start, goal)
print("Path:", path)
```
