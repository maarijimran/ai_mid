```python
from collections import deque
import itertools

city = {
    'HQ': ['Plaza', 'Market'],
    'Plaza': ['HQ', 'School'],
    'Market': ['HQ', 'Hospital'],
    'School': ['Plaza', 'Park'],
    'Hospital': ['Market'],
    'Park': ['School']
}

injured = ['School', 'Hospital', 'Park']
```

```python
def bfs(start, goal):
    visited = set()
    queue = deque([(start, [start])])

    while queue:
        node, path = queue.pop()

        if node == goal:
            return path
        
        if node in visited:
            continue

        visited.add(node)

        for neighbor in city[node]:
            queue.append((neighbor, path + [neighbor]))
```

```python
def calc_utility(order):
    total_utility = 0
    total_moves = 0
    current = 'HQ'
    full_path = []

    for target in order:
        path = bfs(current, target)
        moves = len(path)
        total_moves += moves

        full_path.append((path, target, moves))
        current = target

    total_utility = 10 - total_moves
    return total_utility, full_path
```

```python
def find_path():
    best_utility = float('-inf')
    best_path = None

    for order in itertools.permutations(injured):
        utility, full_path = calc_utility(order)

        if utility > best_utility:
            best_utility = utility
            best_path = (full_path, order)

    print("Best Order Detail : ", best_path[1])
    print()

    for target, moves, path in best_path[0]:
        print(f"Path to {target} : {path}")
        print(f"Moves : {moves}")
        print(f"Utility : {utility}")

    print(f"Total Utility : {best_utility}")
```

```python
find_path()
```
