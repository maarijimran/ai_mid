```python
from queue import PriorityQueue

# Graph with different edge costs
graph = {
    'A': {'B': 2, 'C': 1},
    'B': {'D': 4, 'E': 3},
    'C': {'F': 1, 'G': 5},
    'D': {'H': 2},
    'E': {},
    'F': {'I': 6},
    'G': {},
    'H': {},
    'I': {}
}

# Heuristic function (estimated cost to reach goal 'I')
heuristic = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 4,
    'E': 7,
    'F': 3,
    'G': 6,
    'H': 2,
    'I': 0  # Goal node
}

def a_star(start, goal):
    visited = set()
    pq = PriorityQueue()
    
    pq.put((heuristic[start], start))  # f(n), node
    track = {start: None}
    g_cost = {start: 0}  # cost from start

    while not pq.empty():
        f, node = pq.get()

        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = track[node]
            path.reverse()
            return path
        
        if node in visited:
            continue
        
        visited.add(node)

        for neighbor, cost in graph[node].items():
            new_g = g_cost[node] + cost

            # Only update if better path found
            if neighbor not in g_cost or new_g < g_cost[neighbor]:
                g_cost[neighbor] = new_g
                f_value = new_g + heuristic[neighbor]
                
                pq.put((f_value, neighbor))
                track[neighbor] = node

    return None

path = a_star('A', 'I')
print(path)
```
