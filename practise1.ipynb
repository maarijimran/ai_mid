```python
from collections import deque
import itertools

city = {
    'HQ': ['Plaza', 'Market'],
    'Plaza': ['HQ', 'School'],
    'Market': ['HQ', 'Hospital'],
    'School': ['Plaza', 'Park'],
    'Hospital': ['Market'],
    'Park': ['School']
}

injured = ['School', 'Hospital', 'Park']
```

```python
def bfs(start, goal):
    queue = deque([(start, [start])])
    visited = set()

    while queue:
        node, path = queue.popleft()

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbor in city[node]:
                queue.append((neighbor, path + [neighbor]))

    return None
```

```python
def calc_utility(order):
    totalUtility = 0
    fullPath = []
    current = 'HQ'

    for target in order:
        path = bfs(current, target)
        moves = len(path) - 1
        utility = 10 - moves

        totalUtility += utility

        fullPath.append((target, path, moves, utility))
        current = target

    return totalUtility, fullPath
```

```python
def rescue_agent():
    bestUtility = float('-inf')
    bestPath = None

    for order in itertools.permutations(injured):
        utility, path = calc_utility(order)

        if utility > bestUtility:
            bestUtility = utility
            bestPath = (order, path)

    print("Best Order Detail : ", bestPath[0])
    print()

    for target, path, moves, utility in bestPath[1]:
        print(f"Path to {target} : {path}")
        print(f"Moves : {moves}")
        print(f"Utility : {utility}")

    print(f"Total Utility : {bestUtility}")
    
```

```python
rescue_agent()
```
